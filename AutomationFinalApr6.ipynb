I want this code to:

    1. Take a list of rates and create different models based on this list
    2. Output: File that can be copied into PRISM, CSV documenting the randomly calculated rates

```python
import random
import fileinput
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rates = [15, 20, 21, 17.5, 16, 17, 17.5, 18, 18.5, 19]
```

A range is specified for each individual value in the rates list. 

For example:

    rates_ranges(rates_range, 1):
        rates = [15, 20, 21]
        rates_range = [(14, 15, 16), (19, 20, 21), (20, 21, 22)]

```python
# input: list of rates, empty output list, specified range
# output: list of ranges of the rates +/- the specified range
def rates_ranges(rates_ranges, specified_range):

    updated_ranges = []
    
    for r in rates:
        num_set = (r-specified_range,r,r+specified_range)
        updated_ranges.append(num_set)
    
    return updated_ranges
```

```python
range1 = rates_ranges(rates, 1)
range2 = rates_ranges(rates, 2)
range3 = rates_ranges(rates, 0.5)
range4 = rates_ranges(rates, 0.75)
```

```python
def generate_random_nums_full_range(ranges): # 15 -> [14,16]
        random_nums = []
        for i in range(len(ranges)):
            min_val, max_val = min(ranges[i]), max(ranges[i])
            random_num = random.uniform(min_val, max_val)
            random_nums.append(random_num)
        return random_nums
def generate_random_nums_lower_range(ranges): # 15 -> [14, 15)
        random_nums = []
        for i in range(len(ranges)):
            min_val = min(ranges[i])
            random_num = random.uniform(min_val, ranges[i][1])
            random_nums.append(random_num)
        return random_nums
def generate_random_nums_upper_range(ranges): # 15 -> [15, 16]
        random_nums = []
        for i in range(len(ranges)):
            max_val = max(ranges[i])
            random_num = random.uniform(ranges[i][1], max_val)
            random_nums.append(random_num)
        return random_nums
```

```python
# Generate 10 models in full range +/- 1
range_1 = []
range_1_lower = []
range_1_upper = []
for i in range(10):
    range_1.append(generate_random_nums_full_range(range1))
    range_1_lower.append(generate_random_nums_lower_range(range1))
    range_1_upper.append(generate_random_nums_upper_range(range1))
# Generate 10 models in full range +/- 2
range_2 = []
range_2_lower = []
range_2_upper = []
for i in range(10):
    range_2.append(generate_random_nums_full_range(range2))
    range_2_lower.append(generate_random_nums_lower_range(range2))
    range_2_upper.append(generate_random_nums_upper_range(range2))
# Generate 10 models in full range +/- 0.5
range_05 = []
range_05_lower = []
range_05_upper = []
for i in range(10):
    range_05.append(generate_random_nums_full_range(range3))
    range_05_lower.append(generate_random_nums_lower_range(range3))
    range_05_upper.append(generate_random_nums_upper_range(range3))
# Generate 10 models in full range +/- 0.75
range_075 = []
range_075_lower = []
range_075_upper = []
for i in range(10):
    range_075.append(generate_random_nums_full_range(range4))
    range_075_lower.append(generate_random_nums_lower_range(range4))
    range_075_upper.append(generate_random_nums_upper_range(range4))
```

```python
import graphviz

def draw_tree(tree, rates):
    dot = graphviz.Digraph()
    i = 1
    for node, children in tree.items():
        for child in children:
            if child in rates:
                edge_label = "r" + str(i) + " = " + str(rates[child])
                dot.edge(node, child, label=edge_label)
                i+=1
            else:
                dot.edge(node, child)
    return dot
```

```python
# N = 6
tree = {'at_so_oe_lc_nt_ab': ['nt_ab', 'at_so_oe_lc'],
        'nt_ab': ['nt','ab'],
        'at_so_oe_lc': ['at_so_oe', 'lc'],
        'at_so_oe': ['at_so','oe'],
        'at_so': ['at','so']}

# 'nt_ab': [min(range1[0]),max(range1[0])], 
# Example edge rates
rates = {'nt_ab': random1[0], 
         'at_so_oe_lc': random1[1], 
         'nt': random1[2], 
         'ab': random1[3], 
         'at_so_oe': random1[4], 
         'lc': random1[5],
         'at_so': random1[6],
         'oe': random1[7],
         'at': random1[8],
         'so': random1[9]}

# Draw the tree diagram
dot = draw_tree(tree, rates)

# Render and display the diagram
dot.render('tree', format='png', view=True)
```
